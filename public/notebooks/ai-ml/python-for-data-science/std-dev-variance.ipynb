# Standard Deviation and Variance

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

dataMean = 100.0
stdDev = 50.0
numberOfDataPoints = 10000

incomes = np.random.normal(dataMean, stdDev, numberOfDataPoints)

plt.hist(incomes, 50)
plt.show()
```

```python
incomes.std()
```

```python
incomes.var()
```
