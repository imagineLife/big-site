# Data Distributions

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, expon, binom, poisson
```

## Uniform Distribution
When data is distributed "evenly".

```python
minData = -10.0
maxData = 10
dataPoints = 100000
normalDistVals = np.random.uniform(minData, maxData, dataPoints)
plt.hist(normalDistVals, 50)
plt.show()
```

## Normal / Gaussian
A "bell curve" distribution

```python
minBell = -3
maxBell = 3
valueSpacing = 0.001
x = np.arange(minBell,maxBell,valueSpacing)
plt.plot(x, norm.pdf(x))
```

## Exponential PDF
Like a "hockey stick"

```python
x = np.arange(0, 10, 0.001)
plt.plot(x, expon.pdf(x))
```

## Binomial Probability Mass Function

```python
n, p = 10, 0.5
x = np.arange(0, 10, 0.001)
plt.plot(x, binom.pmf(x, n, p))
```

## Poisson Probability Mass Function

```python
mu = 500
poissonStart = 400
poissonStop = 600
poissonStep = 0.5
x = np.arange(poissonStart, poissonStop, poissonStep)
plt.plot(x, poisson.pmf(x, mu))
```
